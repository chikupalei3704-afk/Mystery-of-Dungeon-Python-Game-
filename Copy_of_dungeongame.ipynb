```python
def start_game():
    health = 100
    inventory = []
    level = 1
    print("\nWelcome to the mystery of dungeon!")
    print("Now you have entered a dungeon. There are two paths ahead.")
    path_choice(health, inventory, level)

def path_choice(health, inventory, level):
    print("\n1. Enter the left path")
    print("2. Enter the right path")
    choice = input("Choose (1 or 2): ")
    if choice == "1":
        left_path(health, inventory, level)
    elif choice == "2":
        right_path(health, inventory, level)
    else:
        print("Invalid choice. Try again.")
        path_choice(health, inventory, level)

def left_path(health, inventory, level):
    print("\nYou enter the left cave.")
    print("1. Explore the cave")
    print("2. Don't want to enter the cave")
    choice = input("Choose (1 or 2): ")
    if choice == "1":
        print("\nWhile exploring the cave, you find a black sword from a dead corpse.")
        inventory.append("Black Sword")
        print("Inventory:", inventory)
    elif choice == "2":
        print("\nYou get scared and try to exit, but fall into a trap and lose 20% of your health.")
        health -= 20
        print(f"Your health is now {health}%")
    else:
        print("Invalid choice. Try again.")
        left_path(health, inventory, level)
    continue_explore(health, inventory, level)

def continue_explore(health, inventory, level):
    print("\nWhile exploring, a behemoth approaches you.")
    print("1. Fight the behemoth with the black sword")
    print("2. Fight the behemoth with your bare hands")
    choice = input("Choose (1 or 2): ")
    if choice == "1":
        print("\nYou successfully defeat the behemoth without losing health.")
    elif choice == "2":
        print("\nYou fight with your bare hands and lose 10% health!")
        health -= 10
        print(f"Your health is now {health}%")
    else:
        print("Invalid choice. Try again.")
        continue_explore(health, inventory, level)
    level += 1
    print(f"Your level is now {level}.")
    final_phase(health, inventory, level)

def right_path(health, inventory, level):
    print("\nThere is a broken bridge over a river.")
    print("1. Cross the river by swimming")
    print("2. Cross the river via the bridge")
    choice = input("Choose (1 or 2): ")
    if choice == "1":
        print("\nYou cross the river by swimming but lose 10% health!")
        health -= 10
        print(f"Your health is now {health}%")
    elif choice == "2":
        print("\nThe bridge collapses, and you fall onto sharp rocks. You lose all health! Game Over.")
        play_again()
        return
    else:
        print("Invalid choice. Try again.")
        right_path(health, inventory, level)
    final_phase(health, inventory, level)

def final_phase(health, inventory, level):
    print("\nYou enter a large room with a giant gate.")
    print(f"Your remaining health: {health}%")
    print(f"Your level: {level}")
    print(f"Your inventory: {inventory}")
    print("An ancient silver dragon blocks your path.")
    print("1. Solve the riddle asked by ancient dragon")
    print("2. Fight the dragon")
    choice = input("Choose (1 or 2): ")
    if choice == "1":
        riddle(health, inventory, level)
    elif choice == "2":
        print("\nYou bravely defeat the dragon but lose 20% health and get a magic potion as reward.")
        health -= 20
        inventory.append("Magic Potion")
        print("Inventory:", inventory)
        level += 1
        print(f"Your level is now {level}.")
        play_again()
    else:
        print("Invalid choice. Try again.")
        final_phase(health, inventory, level)

def riddle(health, inventory, level):
    print("\nThe dragon asks: 'I have words but cannot speak. Hold me close if knowledge you seek. What am I?'")
    attempts = 3
    while attempts > 0:
        answer = input("Your answer: ").lower()
        if answer == "book":
            print("\nYou solved the riddl ,win the game and get a magic potion that have the power to heal every injury as reward.")
            inventory.append("Magic Potion")
            print("Inventory:", inventory)
            level += 1
            print(f"Your level is now {level}.")
            play_again()
            return
        else:
            attempts -= 1
            print(f"\nWrong answer! You have {attempts} attempts left.")
    print("The dragon gets angry and attacks you! You lose 20% health.")
    health -= 20
    print(f"Your health is now {health}%")
    play_again()

def play_again():
    choice = input("\nDo you want to play again? (yes/no): ")
    if choice.lower() == "yes":
        start_game()
    else:
        print("Thank you for playing Mystery of Dungeon!")

start_game()
```
